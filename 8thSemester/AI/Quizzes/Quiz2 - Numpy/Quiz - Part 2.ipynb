```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv(r'.\exams.csv')
```

```python
# Select features for clustering
#features = ['Rating', 'Installs', 'Reviews', 'Size', 'Price']
features = ['reading score','math score','writing score']
X = df[features]
print(X.head)
```

```python
X = df[['reading score', 'writing score']]  # Feature (independent variable)
y = df['math score']    # Target (dependent variable)
```

```python

model = LinearRegression().fit(X, y)
y_pred = model.predict(X)
```

```python
print(f"Slope (Coefficient): {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter actual points
ax.scatter(df['reading score'], df['math score'], y, color='blue', label='Actual')

# Scatter predicted points
ax.scatter(df['reading score'], df['math score'], y_pred, color='red', label='Predicted')

# Labels
ax.set_xlabel('Reading Score')
ax.set_ylabel('Math Score')
ax.set_zlabel('Writing Score')
ax.set_title('3D Multiple Regression Plot')
plt.legend()
plt.show()
```
